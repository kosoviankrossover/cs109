```python
def update_count(tweets):
    # remove stop words and count
    tweet_counter = Counter()
    for i in range(tweets.shape[0]):
        # current words
        curr_wrods = tweets.iloc[i,0].split()

        # words to keep
        words = []
        for w in curr_wrods:
            if w not in stop_words:
                words.append(w)
        new = ' '.join(words)
        tweet_counter += Counter(words)
        tweets.iloc[i,0] = new
    return(tweets, tweet_counter)
```

* age-mean: age of the woman (in years, centered around mean)

The women are grouped into 60 districts. The task is to build a classification model that can predict if a given woman uses contraceptives.


### Part A: Bayesian Logistic Regression

The first model we will fit to the contraceptives data is a varying-intercept logistic regression model, where the intercept varies by district.

Prior distribution:

$\beta_{0j} \sim N(\mu_0,\sigma_0^2),$ with $\mu_0 \sim N(0,10000)$ and $\frac{1}{\sigma^2_0} \sim \text{Gamma}(0.1,0.1)$.

$\beta_1 \sim N(0,10000)$, $\beta_2\sim N(0,10000)$, $\beta_3 \sim N(0,10000)$

Model for data:

$Y_{ij} \sim \text{Bernoulli}(p_{ij})$

$\text{logit}\:p_{ij} = \beta_{0j} + \beta_1 \times \text{urban} + \beta_2 \times \text{living.children} + \beta_3 \times \text{age-mean}$,

where $Y_{ij}$ is 1 if woman $i$ in district $j$ uses contraceptives, and 0 otherwise, and where $i \in \{1,...,N\}$ and $j \in \{1,...,J\}$. $N$ is the number of observations in the data, and $J$ is the number of districts. The above notation assumes $N(\mu, \sigma^2)$ is a Normal distribution with mean $\mu$ and variance $\sigma^2$.

As a preprocessing step, map the district number 61 to the number 54 so that the districts are in order.

```python
# load data
d2_train = pd.read_csv('data/dataset_2_train.csv')
d2_test = pd.read_csv('data/dataset_2_test.csv')
```

```python
# some parameters (train)
N = d2_train.shape[0]
J = len(set(d2_train.district.values))
dvals = set(d2_train.district.values)

# some parameters (test)
Nt = d2_test.shape[0]
Jt = len(set(d2_test.district.values))
```

```python
### map district number 61 to number 54
col_to_i = {}
for i, col in enumerate(d2_train.columns.values):
    col_to_i[col] = i

# find masks
msk_train_61 = d2_train.district.values == 61
msk_test_61 = d2_test.district.values == 61

# # map
d2_train.iloc[msk_train_61, col_to_i['district']] = 54
d2_test.iloc[msk_test_61, col_to_i['district']] = 54
```

To verify the procedure, simulate binary response data assuming the following parameter values (and using the existing features and district information from the training data):

$\mu_{\beta_0} = 2$

$\sigma^2_{\beta_0} = 1$

$\beta_0 \sim N(\mu_{\beta_0},\sigma_{\beta_0}^2)$

$\beta_1 = 4$

$\beta_2 = -3$

$\beta_3 = -2$

```python
# display to look at columns 
display(d2_train.head())
```

```python
#### SIMULATE DATA

# set seed
np.random.seed(1234)

# hypers
mu_beta0 = 2
sigma2_beta0 = 1

# betas
beta0 = np.random.normal(loc= mu_beta0, scale= np.sqrt(sigma2_beta0), size= J)

# betas
beta1 = 4
beta2 = -3
beta3 = -2
betas = np.array([beta1, beta2, beta3])

# index vals for the non-beta0 coeffs
start = 1
end = -1

# ysimulated
ysim = np.zeros(N)-1

for i in range(N):
    # district value
    dval = d2_train.iloc[i,0]
    # convert to j
    j = dval-1
    
    # linear predictor
    lin_pred = beta0[j] + np.dot(betas, d2_train.iloc[i,start:end].values)
    p = expit(lin_pred)
    ysim[i] = np.random.binomial(n=1, p=p, size=1)
```

Fit the varying-intercept model specified above to your simulated data.

```python
# set seed
np.random.seed(5555)

# model string
model_1_string = '''
model {

# prior params
mu_0 ~ dnorm(0,1/10000)
inv_sigma2_0 ~ dgamma(0.1,0.1)
        
# non-beta0 coeffs
beta1 ~ dnorm(0,1/10000)
beta2 ~ dnorm(0,1/10000)
beta3 ~ dnorm(0,1/10000)

# beta0
for (j in 1:J) {
    beta0[j] ~ dnorm(mu_0, inv_sigma2_0)
}

# observations
for (i in 1:N) {
    logit(p[i]) = (beta0[district[i]] + beta1*X[i,1] + beta2*X[i,2] + beta3*X[i,3])
    y[i] ~ dbern(p[i])
}
}
'''
# build model, add data
observed_vals_1 = {'y': ysim, 'N': N, 'J': J, 
                 'district': d2_train.district.values, 
                 'X': d2_train.iloc[:, start:end]}
                 
n_chains = 4
model_1 = pyjags.Model(model_1_string, data= observed_vals_1, chains= n_chains)
n_burnin = 500
burnin = model_1.sample(500) 
n_sim = 10000
model_1_samples = model_1.sample(n_sim)
```

```python
# save data
def save_samps(file_name, samples):
    file_name = file_name+'.pickle'
    pickle_out = open(file_name, 'wb')
    pickle.dump(samples, pickle_out)
    pickle_out.close()
    
# load data
def load_samps(file_name):
    pickle_in = open(file_name, 'rb')
    samples = pickle.load(pickle_in)
    return(samples)
```

Plot the trace plots of the MCMC sampler for the parameters $\mu_{\beta_0}, \sigma_{\beta_0}, \beta_1, \beta_2, \beta_3$. Does it look like the samplers converged?

```python
def plot_trace(chain, title=''):
    fig, ax = plt.subplots(1,1, figsize=(12,6))
    
    n_sim = chain.shape[0]
    n_chains = chain.shape[1]
    
    for cur_chain in range(n_chains):
        cur_label = "Chain {}".format(cur_chain)
        ax.plot(range(n_sim), chain[:,cur_chain], label=cur_label, alpha=0.4)
    ax.legend()
    ax.set_xlabel('Iteration', fontsize=12)
    ax.set_title(title, fontsize=16)
    
    # Hide the right and top spines
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
```

```python
plot_trace(model_1_samples['mu_0'].reshape(n_sim, n_chains), title=r'$\mu_{\beta_0}$')
plot_trace(np.sqrt(np.reciprocal(model_1_samples['inv_sigma2_0'].reshape(n_sim, n_chains))), title=r'$\sigma_{\beta_0}$')
```

```python
plot_trace(model_1_samples['beta1'].reshape(n_sim, n_chains), title=r'$\beta_1$')
plot_trace(model_1_samples['beta2'].reshape(n_sim, n_chains), title=r'$\beta_2$')
plot_trace(model_1_samples['beta3'].reshape(n_sim, n_chains), title=r'$\beta_3$')
```

```python
# set seed
np.random.seed(3834)

# model string
model_string_3 = '''
model {

# prior params
mu_0 ~ dnorm(0,1/10000)
inv_sigma2_0 ~ dgamma(0.1,0.1)
inv_sigma2_1 ~ dgamma(0.1,0.1)
inv_sigma2_2 ~ dgamma(0.1,0.1)
inv_sigma2_3 ~ dgamma(0.1,0.1)
        
# betas
for (j in 1:J) {
    beta0[j] ~ dnorm(mu_0, inv_sigma2_0)
    beta1[j] ~ dnorm(0, inv_sigma2_1)
    beta2[j] ~ dnorm(0, inv_sigma2_2)
    beta3[j] ~ dnorm(0, inv_sigma2_3)
}

# observations
for (i in 1:N) {
    logit(p[i]) = beta0[district[i]] + beta1[district[i]]*X[i,1] + beta2[district[i]]*X[i,2] + beta3[district[i]]*X[i,3]
    y[i] ~ dbern(p[i])
}
}
'''

observed_vals_3 = {'y': d2_train.contraceptive_use.values, 'N': N, 'J': J, 
                 'district': d2_train.district.values, 
                 'X': d2_train.iloc[:, start:end]}
                 
n_chains = 4
model_3 = pyjags.Model(model_string_3, data= observed_vals_3, chains= n_chains)
n_burnin = 500
burnin = model_3.sample(500) 
n_sim = 10000
model_3_samples = model_3.sample(n_sim)
```

From the above trace plots, we can see a number of things. First, we see that the sign and the magnitude of the values for beta0 have flipped and decreased, respectively. Second, we see that the distribution of the third coefficient, beta3, is centered around zero with comparatively small variance; most likely, this coefficient is 0 and provides no information - intuitively, contraceptive is used across ages. Third, it looks like beta0 bet1, and beta2 do have an impact; beta2 appears to have the least impact.
